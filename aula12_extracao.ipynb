````python
people = dataset["results"].copy()
len(people)
````

````python
while dataset["next"] is not None:
  print("Requisição para página", dataset["next"])
  req = requests.get(dataset["next"])
  dataset = req.json()
  people.extend(dataset["results"])
````

- Fez requisição da primeira página
- Identificou como é feita a paginação
- Criou uma lista nova, `people`, e atribuiu as pessoas da primeira requisição
- Enquanto a requisição para a próxima página não for nula:
  - Fez uma nova requisição para a próxima página (`dataset["next"]`)
  - Atribuiu a `dataset` o `json()` da requisição
  - Adicionei a `people` o retorno de `dataset["results"]`

``` json
{
  "url": "xx",
  "startAt": 0,
  "maxResults": 50,
  "total": 180,
  "results": [<50 dados>]
}
```

````python
url = "xpto"
itens = []
start = 0
dataset = requests.get(url).json()

while len(itens) < dataset["total"]:
  req_url = url + "?startAt=" + str(start) # xpto?startAt=0
  print("Requisitando", req_url)
  req = requests.get(req_url)
  dataset = req.json()
  itens.extend(dataset["results"])
  start += 50

"""
Requisitando xpto?startAt=0
Requisitando xpto?startAt=50
Requisitando xpto?startAt=100
Requisitando xpto?startAt=150
"""
````

````python
def salvar_json(conteudo, caminho):
  with open(caminho, "w") as arquivo:
    json.dump(conteudo, arquivo, indent=4)
````

````python
salvar_json(people, "people.json")
````

````python
def ler_json(caminho):
  with open(caminho, "r") as arquivo:
    dados = json.load(arquivo)
  return dados

pessoas = ler_json("people.json")
````

````python
req = requests.get("https://swapi.dev/api/")
req.json()
````

````python
def busca_dados(url):
  print("Requisição para página", url)
  r = requests.get(url)
  dataset = r.json()
  resource = dataset["results"].copy()
  while dataset["next"] is not None:
    print("Requisição para página", dataset["next"])
    r = requests.get(dataset["next"])
    dataset = r.json()
    resource.extend(dataset["results"])
  
  return resource
````

````python
people = busca_dados("https://swapi.dev/api/people/")
salvar_json(people, "people.json")
````
